# Dics Cogs API

```python
import discogs_client

DISCOGS_KEY = os.environ.get('DISCOGS_ACCESS_TOKEN')
d = discogs_client.Client('DadRecords',user_token=DISCOGS_KEY)
```

```python
album_name = 'mask of smiles'
cat_id = r"st-17146"
country = 'canada'


```

```python
import pandas as pd
results = d.search(album_name, type='release')

results_list = [result.data for result in results]
df = pd.DataFrame(results_list)

```

```python
df
```

```python

# Filter by catno and country if both variables are not empty
if cat_id!='' and country!='':
    filtered_df = df[df['catno'].str.contains(cat_id, case=False) & df['country'].str.contains(country, case=False)]
    print(filtered_df)
# Filter by catno if country is empty
elif cat_id!='':
    filtered_df = df[df['catno'].str.contains(cat_id, case=False)]
    print(filtered_df)
# Filter by country if cat_id is empty
elif country!='':
    filtered_df = df[df['country'].str.contains(country, case=False)]
    print(filtered_df)

else:
    print('Please supply at least catno or country')
```
